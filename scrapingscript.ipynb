```python
import requests
from bs4 import BeautifulSoup

def scrape_page(url):
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:87.0) Gecko/20100101 Firefox/87.0'}
    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()  # Cela déclenchera une exception si le statut n'est pas 200
    except requests.HTTPError as http_err:
        print(f"Une erreur HTTP est survenue : {http_err}")
        return None
    except Exception as err:
        print(f"Une autre erreur est survenue : {err}")
        return None

    # Si la réponse est réussie, poursuivre le scraping
    soup = BeautifulSoup(response.text, 'html.parser')

    # Adaptez les sélecteurs ci-dessous selon la structure exacte du site
    expressions = []
    for row in soup.select('table tr')[1:]:  # Modifier pour sélectionner les lignes correctes, sauter l'en-tête
        cells = row.find_all('td')
        if len(cells) >= 1:  # Assurez-vous qu'il y a assez de cellules
            english = cells[0].get_text(strip=True)
            french = cells[2].get_text(strip=True)
            expressions.append((english, french))

    return expressions

# URL de la page à scraper
url = "https://www.expressio.fr/expressions-idiomatiques-en-anglais"

# Appel de la fonction et affichage des résultats
expressions = scrape_page(url)
if expressions:
    for english, french in expressions:
        print(f"{english} -> {french}")
else:
    print("Aucune expression trouvée ou une erreur s'est produite.")
```

```python
import requests
from bs4 import BeautifulSoup
import csv

def scrape_page(url):
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:87.0) Gecko/20100101 Firefox/87.0'}
    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()
    except requests.HTTPError as http_err:
        print(f"Une erreur HTTP est survenue : {http_err}")
        return []
    except Exception as err:
        print(f"Une autre erreur est survenue : {err}")
        return []

    soup = BeautifulSoup(response.text, 'html.parser')
    expressions = []
    for row in soup.select('table tr')[1:]:
        cells = row.find_all('td')
        if len(cells) >= 1:
            english = cells[0].get_text(strip=True)
            french = cells[2].get_text(strip=True)
            expressions.append((english, french))

    return expressions

# Traiter la première page séparément
url = "https://www.expressio.fr/expressions-idiomatiques-en-anglais"
all_expressions = scrape_page(url)

# Continuer avec les autres pages
start_page = 2
end_page = 191

for i in range(start_page, end_page + 1):
    url = f"https://www.expressio.fr/expressions-idiomatiques-en-anglais/page-{i}"
    print(f"Scraping {url}")
    expressions = scrape_page(url)
    all_expressions.extend(expressions)

# Sauvegarder dans un fichier CSV

csv_file_path = 'expressions_idiomatiques_anglais_francais.csv'

with open(csv_file_path, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Expression en anglais', 'Expression en français'])
    writer.writerows(all_expressions)

print(f"Les données ont été enregistrées dans {csv_file_path}")
```
